```python
import pandas as pd
import re
```

```python
dfb = pd.read_csv('Output\Early_Bike_Scraped.csv')
dfc = pd.read_csv('Output\Early_Car_Scraped.csv')
```

```python
dfc.dropna(inplace=True)
dfb.dropna(inplace=True)
dfc.drop_duplicates(subset= ['title', 'price', 'brand', 'type', 'location'], keep='first', inplace=True)
dfb.drop_duplicates(subset= ['title', 'price', 'brand', 'type', 'location'],keep='first', inplace=True)
```

```python
dfc.price = dfc.price.astype(int)
dfc.years = dfc.years.astype(int)
dfb.price = dfb.price.astype(int)
dfb.years = dfb.years.astype(int)
```

```python
check_type = [
    'avanza', 
    'xpander', 
    'brio', 
    'rush', 
    'innova',
    'crv', 
    'hrv', 
    'jazz', 
    'yaris', 
    'beat', 
    'scoopy',
    'vario',
    'mio'
    ]
dfc = dfc[(dfc['title'].str.contains('|'.join(check_type), case=False))]
dfb = dfb[(dfb['title'].str.contains('|'.join(check_type), case=False))]
```

```python
dfc = dfc[(dfc['years'] < 2022) & (dfc['years'] > 2015)]
dfb = dfb[(dfb['years'] < 2022) & (dfb['years'] > 2015)]
```

```python
milage = pd.DataFrame()
```

```python
for i in range(0, 61):
    if i == 60:
        milage.loc[i, 'base_milage'] = '&gt;300.000'
        milage.loc[i, 'upper_milage'] = ''
        milage.loc[i, 'milage_val'] = milage.loc[i, 'base_milage'] + milage.loc[i, 'upper_milage']
        milage.loc[i, 'key'] = str(i+1)
    else:
        milage.loc[i, 'base_milage'] = re.sub(r'(?<!^)(?=(\d{3})+$)', r'.',str(i*5000))
        milage.loc[i, 'upper_milage'] = re.sub(r'(?<!^)(?=(\d{3})+$)', r'.',str((i+1)*5000))
        milage.loc[i, 'milage_val'] = milage.loc[i, 'base_milage'] + '-' + milage.loc[i, 'upper_milage']
        milage.loc[i, 'key'] = str(i+1)
```

```python
milage = milage.iloc[:,2:4]
```

```python
milage_dict = {}
for i in range(0,61):
    milage_dict.setdefault(milage.iloc[i,0], []).append(str(milage.iloc[i,1]))
```

```python
dfb['milage'].replace(to_replace =r'.Km|.km', value = '', regex = True, inplace=True)
dfc['milage'].replace(to_replace =r'.Km|.km', value = '', regex = True, inplace=True)
```

```python
check_milage = list(milage_dict.keys())
pat = '|'.join(r"\b{}\b".format(x) for x in check_milage)
dfc = dfc[dfc['milage'].str.contains(pat)]
dfb = dfb[dfb['milage'].str.contains(pat)]
```

```python
dfc.reset_index(drop=True, inplace=True)
dfb.reset_index(drop=True, inplace=True)
```

```python
for i in range(0, len(dfc.index)):
    current = dfc.loc[i, 'milage']
    dfc.loc[i, 'milage_class'] = milage_dict[current][0]

for i in range(0, len(dfb.index)):
    current = dfb.loc[i, 'milage']
    dfb.loc[i, 'milage_class'] = milage_dict[current][0]
```

```python
cols = list(dfc.columns)
a, b = cols.index('transmission'), cols.index('milage_class')
cols[b], cols[a] = cols[a], cols[b]
dfc = dfc[cols]
```

```python
dfc = dfc.sort_values(['brand','type','years','milage_class','price'], ascending=[False,False,False,False,False])
dfb = dfb.sort_values(['brand','type','years','milage_class','price'], ascending=[False,False,False,False,False])
```

```python
type_dict = {
    'toyota avanza' : 'avanza',
    'toyota rush' : 'rush',
    'toyota innova' : 'innova',
    'toyota yaris' : 'yaris',
    'mitsubishi xpander' : 'xpander',
    'honda brio' : 'brio',
    'honda crv' : 'cr-v',
    'honda hrv' : 'hr-v',
    'honda jazz' : 'jazz'
}
```

```python
def call_type(name):
    type = type_dict[name]
    return type
```

```python
dfc.type = dfc.title.apply(call_type)
```

```python
dfc = dfc[(dfc['price'] < 800000000) & (dfc['price'] > 50000000)]
dfb = dfb[(dfb.price < 25000000) & (dfb.price > 3000000)]
```
