```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
dfc = pd.read_csv('Output\Mid_Car_Cleaned.csv')
dfb = pd.read_csv('Output\Mid_Bike_Cleaned.csv')
```

```python
def print_full(x):
    pd.set_option('display.max_rows', None)
    pd.set_option('display.max_columns', None)
    pd.set_option('display.width', 2000)
    pd.set_option('display.float_format', '{:20,.2f}'.format)
    pd.set_option('display.max_colwidth', None)
    print(x)
    pd.reset_option('display.max_rows')
    pd.reset_option('display.max_columns')
    pd.reset_option('display.width')
    pd.reset_option('display.float_format')
    pd.reset_option('display.max_colwidth')
```

```python
mb =dfb.groupby(['type', 'years']).agg({'price': ['median','mean', 'min', 'max']}).sort_values(by=['type','years'], ascending=False)
mb.columns = ['price_median','price_mean', 'price_min', 'price_max']
mb = mb.reset_index()
mb = pd.DataFrame(data=mb)
```

```python
print_full(mb)
```

```python
mc =dfc.groupby(['type', 'years']).agg({'price': ['median','mean', 'min', 'max']}).sort_values(by=['type','years'], ascending=False)
mc.columns = ['price_median','price_mean', 'price_min', 'price_max']
mc = mc.reset_index()
mc = pd.DataFrame(data=mc)
```

```python
print_full(mc)
```

```python
mc['index'] = mc.price_median / mc.groupby('type').price_median.transform(np.max)
```

```python
print_full(mc)
```

```python
mb['index'] = mb.price_median / mb.groupby('type').price_median.transform(np.max)
```

```python
print_full(mb)
```

```python
mb.set_index('years', inplace=True)
mb.groupby('type')['index'].plot(legend=True, figsize=(20,10))
mb = mb.reset_index()
```

```python
mc.set_index('years', inplace=True)
mc.groupby('type')['index'].plot(legend=True, figsize=(20,10))
mc = mc.reset_index()
```

```python
mb['pct_ch'] = mb.groupby('type')['price_mean'].pct_change()
```

```python
print_full(mb)
```

```python
mc['pct_ch'] = mc.groupby('type')['price_mean'].pct_change()
```

```python
print_full(mc)
```

```python
mb.set_index('years', inplace=True)
mb.groupby('type')['pct_ch'].plot(legend=True, figsize=(20,10))
mb = mb.reset_index()
```

```python
mc.set_index('years', inplace=True)
mc.groupby('type')['pct_ch'].plot(legend=True, figsize=(20,10))
mc = mc.reset_index()
```

```python
mbx = mb.groupby(['years']).agg({'pct_ch': ['median','mean', 'min', 'max']}).sort_values(by=['years'], ascending=False)
mbx.columns = ['pct_median','pct_mean', 'pct_min', 'pct_max']
mbx = mbx.reset_index()
mbx = pd.DataFrame(data=mbx)
```

```python
mbx = mbx.dropna()
```

```python
mbx.set_index('years', inplace=True)
mbx['pct_mean'].plot(legend=True, figsize=(20,10))
mbx = mbx.reset_index()
```

```python
mcx = mc.groupby(['year']).agg({'pct_ch': ['median','mean', 'min', 'max']}).sort_values(by=['year'], ascending=False)
mcx.columns = ['pct_median','pct_mean', 'pct_min', 'pct_max']
mcx = mcx.reset_index()
mcx = pd.DataFrame(data=mcx)
```

```python
mcx = mcx.dropna()
```

```python
mcx.set_index('year', inplace=True)
mcx['pct_mean'].plot(legend=True, figsize=(20,10))
mcx = mcx.reset_index()
```

```python
mby = mb.groupby(['years']).agg({'index': ['median','mean', 'min', 'max']}).sort_values(by=['years'], ascending=False)
mby.columns = ['in_median','in_mean', 'in_min', 'in_max']
mby = mby.reset_index()
mby = pd.DataFrame(data=mby)
```

```python
mby = mby.dropna()
```

```python
mby.set_index('years', inplace=True)
mby['in_mean'].plot(legend=True, figsize=(20,10))
mby = mby.reset_index()
```

```python
mcy =mc.groupby(['year']).agg({'index': ['median','mean', 'min', 'max']}).sort_values(by=['year'], ascending=False)
mcy.columns = ['in_median','in_mean', 'in_min', 'in_max']
mcy = mcy.reset_index()
mcy = pd.DataFrame(data=mcy)
```

```python
mcy = mcy.dropna()
```

```python
mcy.set_index('year', inplace=True)
mcy['in_mean'].plot(legend=True, figsize=(20,10))
mcy = mcy.reset_index()
```
